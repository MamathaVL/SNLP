```python
import nltk
text = open("LoR_Sample.txt")
data = text.read()
```

```python
#Define grammar using regular expressions
grammar = ('''
    NP: {<DT>?<JJ>*<NN>} # NP
    ''')
```

```python
chunkParser = nltk.RegexpParser(grammar)
tagged = nltk.pos_tag(nltk.word_tokenize(data))
tagged
```

```python
tree = chunkParser.parse(tagged)
```

```python
for subtree in tree.subtrees():
    print(subtree)
```

```python
tree.draw()
```

```python
#Import library
import spacy
```

```python
#Loading the default model english-core-web
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(u'Ali Hassan Kuban said that Apple Inc. will buy Google in May 2018.')
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
