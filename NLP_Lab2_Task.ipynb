```python
#Task : Basic Text Processing Pipeline
import nltk
sent = "Become an expert in NLP!"
words = nltk.word_tokenize(sent)
print(words)
```

```python
texts = "Machine learning is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed. Machine learning focuses on the development of computer programs that can access data and use it learn for themselves"
for text in texts:
    sentences = nltk.sent_tokenize(texts)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        print(words)
```

```python
tagged = nltk.pos_tag(words)
print(tagged)
```

```python
#Extract Text from Corpus - Brown
import nltk
from nltk.corpus import brown
```

```python
#Access Data of the Corpus - Brown with category 'news'- 100 words
text=brown.words(categories='news')[:100]
```

```python
#Tokenization
texts = ' '.join([str(item) for item in text ])
for text in texts:
    sentences = nltk.sent_tokenize(texts)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        print(words)
```

```python
#POS_Tagging
nltk.pos_tag(words)
```

```python
#Determine the stems in those words using PorterStemmer
import nltk
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()

for w in words:
       print("Actual: %s     Stem: %s"  %(w,porter_stemmer.stem(w)))
```

```python
#Determine the stems in those words using LancasterStemmer
from nltk.stem import LancasterStemmer
stemmerLan = LancasterStemmer()

for w in words:
       print("Actual: %s       Stem: %s"  %(w,stemmerLan.stem(w)))
```

```python
#Determine the stems in those words using SnowballStemmer
from nltk.stem import SnowballStemmer
stemmer= SnowballStemmer('english')

for w in words:
       print("Actual: %s      Stem: %s"  %(w,stemmer.stem(w)))
```
